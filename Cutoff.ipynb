```python
#List of all categories
categoryS=['GOPENS', 'GSCS', 'GSTS', 'GVJS', 'GNT1S', 'GNT2S', 'GNT3S', 'GOBCS', 'GSEBCS', 'LOPENS', 'LSCS', 'LSTS', 'LVJS', 'LNT1S', 'LNT2S', 'LNT3S', 'LOBCS', 'LSEBCS', 'PWDOPENS', 'PWDSCS', 'PWDOBCS', 'DEFOPENS', 'EWS', 'TFWS']
categoryH=[]
categoryO=[]
for i in categoryS:
    h,o= 'H','O'
    k= i[:-1]+h
    m= i[:-1]+o
    categoryH.append(k)
    categoryO.append(m)
category = categoryS + categoryH + categoryO
print('Category = ',category)
```
